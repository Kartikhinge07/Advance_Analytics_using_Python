```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
sal = pd.read_csv(r"E:\Adv Analytics\Datasets\Exp_Salaries.csv")
sal
```

```python
# in x all attribute will show expect 'salary'
# in y only salary
x = sal.drop('Salary', axis=1)
y=sal['Salary']
x = pd.get_dummies(x,drop_first= True)
x
```

```python
lr= LinearRegression()
lr.fit(x,y)
print("Slope=", lr.coef_)
print("Intercept=",lr.intercept_)
# Here slope will show for all attribute    
```

```python
y_pred_trn = lr.predict(x)
print(mean_absolute_error(y, y_pred_trn))
print(mean_squared_error(y, y_pred_trn))
print(r2_score(y, y_pred_trn))
```

```python
dum_house = pd.get_dummies(housing,drop_first = True)
train,test = train_test_split(dum_house , test_size=0.3,random_state=23)
x_train = train.drop('price' ,axis=1)
y_train = train['price']
x_train
# pd.getdummies will fill the categorical data with the binary values
# in x_train  all value expect price wil dispaly
```

```python
y_train
# in this only price will display
```

```python
lr= LinearRegression()
lr.fit(x_train,y_train)
print("Intercept", lr.intercept_)
print("Slopes",lr.coef_)
```

```python
x_test = test.drop('price',axis=1)
y_pred = lr.predict(x_test) #it will predict price by all other independent attributes
y_test = test['price'] #it will take original price that store in test at time of split 
```

```python
# here we are comparing the value of predicted price (y_pred) and original price(y_test)
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
