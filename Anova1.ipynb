```python
import pandas as pd 
from statsmodels.stats.anova import anova_lm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
agr = pd.read_csv(r"E:\Adv Analytics\Datasets\Yield.csv")
agr
```

```python
agr.groupby('Treatments').mean()
```

```python
#H0 = mean of treatments(1,2,3,4) have no difference
#H1 = mean of treatments(1,2,3,4) have  difference
agrYield = ols('Yield ~ Treatments', data=agr).fit()
table = anova_lm(agrYield, typ=2)
print(table)
```

```python
### Post hoc test : Tukey's HSD Test
compare = pairwise_tukeyhsd(agr['Yield'], 
                            agr['Treatments'], alpha=0.05)
pd.DataFrame(compare._results_table.data)

#as we reject H0 ,so we accept H1 that mean there is difference in there mean
#To check mean between treatments(1,2,3,4) we use this method and check which treatment have difference ( 1 or 2 , 2 or 3) like this 
```

```python
ols_plant = ols('weight ~ group', data=plant).fit()
table= anova_lm(ols_plant,typ=2)
print(table)
```
