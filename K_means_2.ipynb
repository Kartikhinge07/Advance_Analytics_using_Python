```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 
```

```python
nutrients = pd.read_csv(r"E:\Adv Analytics\Datasets\nutrient.csv" , index_col=0)
nutrients
```


### Scaling the Data

```python
scaler = MinMaxScaler()
nutrients_scl = scaler.fit_transform(nutrients)
nutrients_scl = pd.DataFrame(nutrients_scl,
                        columns=nutrients.columns,
                        index=nutrients.index)
```

### Examining the number of clusters

```python
wss = []
nos = [2,3,4,5,6,7,8,9,10]
for i in nos:
    km = KMeans(n_clusters=i, random_state=23)
    km.fit(nutrients_scl)
    wss.append(km.inertia_)
    
plt.plot(nos,wss)
plt.scatter(nos,wss,color="red")
plt.title("Scree Plot")
plt.xlabel("no of cluster")
plt.ylabel("WSS")
plt.show()
```

############## US ARREST #####################

```python
USArrest = pd.read_csv(r"E:\Adv Analytics\Datasets\USArrests.csv" , index_col=0)
USArrest
```

```python
scaler = MinMaxScaler()
USArrest_scl = scaler.fit_transform(USArrest)
USArrest_scl = pd.DataFrame(USArrest_scl,
                        columns=USArrest.columns,
                        index=USArrest.index)
```

```python
wss = []
nos = [2,3,4,5,6,7,8,9,10]
for i in nos:
    km = KMeans(n_clusters=i, random_state=23)
    km.fit(USArrest_scl)
    wss.append(km.inertia_)
    
plt.plot(nos,wss)
plt.scatter(nos,wss,color="red")
plt.title("Scree Plot")
plt.xlabel("no of cluster")
plt.ylabel("WSS")
plt.show()
```

### Clustering the data

```python
km = KMeans(n_clusters=4)
km.fit(USArrest_scl)
print(km.inertia_)
print(km.labels_)

clust_USArrest = USArrest.copy()
clust_USArrest['Cluster'] = km.labels_
clust_USArrest.sort_values('Cluster')


```

```python
clust_USArrest.groupby('Cluster').mean()
```
