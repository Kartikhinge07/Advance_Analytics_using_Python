```python
import pandas as pd
from scipy.stats import ttest_rel
```

A group of seven patients of rheumatic heart disease with distention of abdomen due to ascites, affecting breathing capacity were treated. Can we say that treatment has improved breathing capacity? Data is in file Rheumatic.csv

```python
rhuem = pd.read_csv(r"E:\Adv Analytics\Datasets\Rheumatic.csv")
rhuem
```

```python
result = ttest_rel(rhuem['Before'], rhuem['After'],
          alternative='less')

test_stat = result[0]
p_value = result[1]
print("Test Statistic =", test_stat)
print("P - Value =", p_value)
```

WE HAVE TO ALWAYS USE BEFORE - AFTER . AFTER - BEFORE WILL GIVE ANSWER BUT IT IS NOT RECOMMAND TO USE IT.

```python

result = ttest_rel(rhuem['After'], rhuem['Before'],
          alternative='greater')

test_stat = result[0]
p_value = result[1]
print("Test Statistic =", test_stat)
print("P - Value =", p_value)
```
